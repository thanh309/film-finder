## **Collaborative-Filtering Recommender System**

### **Import necessary libraries**

```python
import pandas as pd 
import numpy as np 
from scipy import sparse
from sklearn.metrics.pairwise import cosine_similarity
```

### **A quick look at the dataset**

```python
test_data = pd.read_csv('testing_data.csv')
train_data = pd.read_csv('training_data.csv')

test_data.head(5)
```

```python
train_data.head(5)
```

```python
ratings_train = train_data.values
ratings_test = test_data.values

# indices start from 0
ratings_train[:, :2] -= 1
ratings_test[:, :2] -= 1
```

```python
n_users = int(np.max(ratings_train[:, 0])) + 1
n_items = int(np.max(ratings_train[:, 1])) + 1
n_users, n_items
```

### **Build the Utility Matrix**

```python
uu_utility_matrix = sparse.coo_matrix((ratings_train[:, 2],
            (ratings_train[:, 1], ratings_train[:, 0])), shape=(n_items, n_users))
uu_utility_matrix = uu_utility_matrix.tocsr()
```

```python
uu_utility_matrix.toarray()
```
